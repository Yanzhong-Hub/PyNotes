```python
import pandas as pd
import json
```

```python
with open('test.json') as f:
    data = json.load(f)

data = pd.DataFrame(data['data'])
data
```

```python
data = data.rename(columns={'tradingDay': 'trade_date'})
data['trade_date'] = pd.to_datetime(data['trade_date']).dt.date
data
```

```python
import matplotlib.plot as plt
hj = data.query('productName == "和聚7期"')
hj.plot(x='trade_date', y='netTotalValue', grid='both')
```
